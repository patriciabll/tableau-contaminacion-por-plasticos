Vamos a mirar las columnas que tienen nulos y su porcentaje. Ordenamos de mayora porcentaje a menor.

```python
df_nulos = pd.DataFrame(df.isnull().sum() / df.shape[0] * 100, columns = ["%_nulos"])
df_nulos = df_nulos.sort_values(by='%_nulos', ascending=False)
df_nulos
```
